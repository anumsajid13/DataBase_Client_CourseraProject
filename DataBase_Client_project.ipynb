```python
try:
    # Step one: Import MySQLConnectionPool

    # Step two: Import Error

    # Step three: Create a pool named pool_a with two connections. Use a try-except block to handle any possible errors.

    # Creating a connection pool with two connections
    pool_a = MySQLConnectionPool(
        pool_name="pool_a",
        pool_size=2,
        host="localhost",
        user="root",
        password="anumsajid@4321",
        database="little_lemon_db1"
    )

    # further operations with the connection pool if no errors occurred
    connection = pool_a.get_connection()
    cursor = connection.cursor()
    
     # SQL CREATE PROCEDURE query for PeakHours
    create_procedure_query1 = """
    CREATE PROCEDURE PeakHours4()
    BEGIN
        SELECT HOUR(BookingSlot) AS BookingHour, COUNT(*) AS TotalBookings
        FROM Bookings
        GROUP BY BookingHour
        ORDER BY TotalBookings DESC;
    END
    """
    
    cursor.execute(create_procedure_query1)
    cursor.callproc("PeakHours4")

    dataset = cursor.fetchall()
  
 

    # Print the sorted data using a for loop.
    for row in dataset:
        print(row)

    # Write a SQL CREATE PROCEDURE query for GuestStatus
    create_procedure_query2 = """
    CREATE PROCEDURE GuestStatus4()
    BEGIN
        SELECT CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) AS GuestName,
               CASE
                   WHEN employees.Role IN ('Manager', 'Assistant Manager') THEN 'Ready to pay'
                   WHEN employees.Role = 'Head Chef' THEN 'Ready to serve'
                   WHEN employees.Role = 'Assistant Chef' THEN 'Preparing Order'
                   WHEN employees.Role = 'Head Waiter' THEN 'Order served'
                   ELSE 'Unknown'
               END AS Status
        FROM Bookings
        LEFT JOIN employees ON Bookings.EmployeeID = Employees.EmployeeID;
    END
    """
    
    cursor.execute(create_procedure_query2)

    # Invoke callproc to call the stored procedure.
    cursor.callproc("GuestStatus4")

    # Fetch the results in a variable called dataset.
    dataset = cursor.fetchall()


    # Print the sorted data using a for loop.
    for row in dataset:
        print(row)

    # Close the connection to return it back to the pool.
    cursor.close()
    connection.close()
    
except Error as e:
    # Handle any possible errors
    print(f"Error: {e}")



```
